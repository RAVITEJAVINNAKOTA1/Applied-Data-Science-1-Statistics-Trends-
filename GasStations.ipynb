```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# @title totalScore
def plot_total_score_histogram(dataframe, bins=20):
    dataframe['totalScore'].plot(kind='hist', bins=bins, title='totalScore')
    plt.gca().spines[['top', 'right']].set_visible(False)
    plt.show()

plot_total_score_histogram(df)
```

```python
# @title Category Name

def plot_category_name_barh(dataframe):
    category_counts = dataframe.groupby('categoryName').size()
    category_counts.plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
    plt.gca().spines[['top', 'right']].set_visible(False)
    plt.show()

plot_category_name_barh(df)
```

```python
# @title reviewsCount

def plot_reviews_count_line(dataframe):

    fig, ax = plt.subplots(figsize=(8, 4))
    dataframe['reviewsCount'].plot(kind='line', ax=ax, marker='o', color='b')
    ax.set_title('reviewsCount')
    ax.set_xlabel('Index')
    ax.set_ylabel('Number of Reviews')
    ax.grid(True)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.show()

plot_reviews_count_line(df)
```

```python
# @title Correlation Map

def plot_correlation_map(dataframe):

    correlation_matrix = dataframe.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Map')
    plt.show()

plot_correlation_map(df)
```
